```python
# === 1. SETUP ===
!pip install --upgrade pip
!pip install PyWavelets
!pip install wfdb

import numpy as np
import matplotlib.pyplot as plt
import wfdb
import random
import os
from scipy.signal import find_peaks
from sklearn.model_selection import train_test_split, GridSearchCV
from sklearn.svm import SVR
from sklearn.multioutput import MultiOutputRegressor
from sklearn.metrics import mean_absolute_error, hamming_loss, f1_score, roc_auc_score
from sklearn.preprocessing import StandardScaler
import pywt
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv1D, MaxPooling1D, GlobalAveragePooling1D, Dense, Dropout
from tensorflow.keras.callbacks import EarlyStopping

# Parameters
FS = 360
SEG_LEN = 5 * FS  # 1800 samples
NUM_SEGMENTS = 10000
SNR_RANGE = (-10, 20)  # Wide range for varying severity levels
RANDOM_SEED = 42
random.seed(RANDOM_SEED)
np.random.seed(RANDOM_SEED)

NOISE_LABELS = ['BW', 'MA', 'EM']
NOISE_RECORDS = {'BW': 'bw', 'MA': 'ma', 'EM': 'em'}
CLEAN_RECORDS = ['100', '101', '102', '103', '105', '107', '108', '109', '111', '112']
```

```python
# === 2. DOWNLOAD DATA ===
if not os.path.exists('mitdb'):
    wfdb.dl_database('mitdb', 'mitdb')
if not os.path.exists('nstdb'):
    wfdb.dl_database('nstdb', 'nstdb')

clean_signals = []
for rec in CLEAN_RECORDS:
    record = wfdb.rdrecord(f'mitdb/{rec}')
    sig = record.p_signal[:, 0].flatten()
    clean_signals.append(sig)

noise_signals = {}
for label, rec in NOISE_RECORDS.items():
    record = wfdb.rdrecord(f'nstdb/{rec}')
    noise_signals[label] = record.p_signal.flatten()
```

```python
# === 3. HELPERS ===
def min_max_normalize(seg: np.ndarray) -> np.ndarray:
    mn, mx = seg.min(), seg.max()
    if mx - mn == 0:
        return np.zeros_like(seg)
    return (seg - mn) / (mx - mn)

def add_noise(clean_seg: np.ndarray, noise_seg: np.ndarray, snr_db: float) -> np.ndarray:
    p_clean = np.mean(clean_seg ** 2)
    p_noise = np.mean(noise_seg ** 2)
    if p_noise == 0:
        return clean_seg
    scale = np.sqrt(p_clean / (p_noise * 10 ** (snr_db / 10.0)))
    return clean_seg + scale * noise_seg

def compute_terma(seg: np.ndarray) -> np.ndarray:
    sq = seg ** 2
    short_win, long_win = 40, 150
    ma_short = np.convolve(sq, np.ones(short_win)/short_win, mode='same')
    ma_long = np.convolve(sq, np.ones(long_win)/long_win, mode='same')
    return ma_short - ma_long

def extract_terma_features(seg: np.ndarray) -> np.ndarray:
    terma_sig = compute_terma(seg)
    mean_val = np.mean(terma_sig)
    std_val = np.std(terma_sig)
    # Robust peak count
    prominence = std_val / 2 if std_val > 0 else 0.01
    num_peaks = len(find_peaks(terma_sig, prominence=prominence)[0])
    return np.array([mean_val, std_val, num_peaks])

def extract_dwt_features(seg: np.ndarray) -> np.ndarray:
    coeffs = pywt.wavedec(seg, 'db4', level=3)
    features = []
    for c in coeffs:
        features.extend([np.std(c), np.sum(c ** 2)])  # std + energy
    return np.array(features)

def extract_features(seg: np.ndarray) -> np.ndarray:
    return np.concatenate([extract_dwt_features(seg), extract_terma_features(seg)])
```

```python
# === 4. CREATE DATASET WITH MIXTURES ===
all_clean_segs_norm = []
for sig in clean_signals:
    for i in range(0, len(sig) - SEG_LEN, SEG_LEN // 2):
        seg = sig[i:i + SEG_LEN]
        all_clean_segs_norm.append(min_max_normalize(seg))

print(f"Available clean segments: {len(all_clean_segs_norm)}")

X = []
y_frac = []  # (N, 3) fractions 0-1

for _ in range(NUM_SEGMENTS):
    clean_norm = random.choice(all_clean_segs_norm)
    clean_raw = clean_norm * 2 - 1
    p_clean = np.mean(clean_raw ** 2)

    noisy_raw = clean_raw.copy()
    added_powers = [0.0] * 3

    num_noises = random.randint(0, 3)
    chosen = random.sample(range(3), num_noises) if num_noises > 0 else []

    for idx in chosen:
        label = NOISE_LABELS[idx]
        start = random.randint(0, len(noise_signals[label]) - SEG_LEN)
        noise_snip = noise_signals[label][start:start + SEG_LEN]
        snr_db = random.uniform(*SNR_RANGE)
        added = add_noise(np.zeros_like(clean_raw), noise_snip, snr_db)  # add at target SNR relative to clean power
        noisy_raw += added
        added_powers[idx] = np.mean(added ** 2)

    p_total = np.mean(noisy_raw ** 2) or 1e-10
    fracs = [ap / p_total for ap in added_powers]

    X.append(min_max_normalize(noisy_raw))
    y_frac.append(fracs)

X = np.array(X)
y_frac = np.array(y_frac)

print(f"Dataset: {X.shape}, Mean severity % per noise: {(y_frac.mean(axis=0)*100).round(1)}")
```

```python
# === 5. SPLIT ===
X_train_val, X_test, y_train_val, y_test = train_test_split(X, y_frac, test_size=0.15, random_state=RANDOM_SEED)
X_train, X_val, y_train, y_val = train_test_split(X_train_val, y_train_val, test_size=0.15, random_state=RANDOM_SEED)
```

```python
# === 6. SVM REGRESSION ===
print("Extracting features for SVM...")
X_train_feat = np.array([extract_features(seg) for seg in X_train])
X_val_feat   = np.array([extract_features(seg) for seg in X_val])
X_test_feat  = np.array([extract_features(seg) for seg in X_test])

scaler = StandardScaler()
X_train_feat = scaler.fit_transform(X_train_feat)
X_val_feat   = scaler.transform(X_val_feat)
X_test_feat  = scaler.transform(X_test_feat)

svr = MultiOutputRegressor(SVR(kernel='rbf', C=10.0, gamma='scale'))  # Small grid-recommended values
svr.fit(X_train_feat, y_train)

y_pred_svm_frac = svr.predict(X_test_feat)
y_pred_svm = y_pred_svm_frac * 100
y_test_perc = y_test * 100

mae_svm = [mean_absolute_error(y_test_perc[:, i], y_pred_svm[:, i]) for i in range(3)]
print(f"SVM MAE (%): BW={mae_svm[0]:.2f}, MA={mae_svm[1]:.2f}, EM={mae_svm[2]:.2f} | Avg={np.mean(mae_svm):.2f}")
```

```python
# === 7. CNN REGRESSION ===
X_train_cnn = X_train[..., np.newaxis]
X_val_cnn   = X_val[..., np.newaxis]
X_test_cnn  = X_test[..., np.newaxis]

model = Sequential([
    Conv1D(32, 11, activation='relu', input_shape=(SEG_LEN, 1)),
    MaxPooling1D(3),
    Conv1D(64, 7, activation='relu'),
    MaxPooling1D(3),
    Conv1D(128, 5, activation='relu'),
    MaxPooling1D(3),
    GlobalAveragePooling1D(),
    Dropout(0.5),
    Dense(32, activation='relu'),
    Dense(3, activation='sigmoid')  # 0-1 fractions
])

model.compile(optimizer='adam', loss='mse', metrics=['mae'])
model.summary()

early_stop = EarlyStopping(patience=5, restore_best_weights=True)
history = model.fit(X_train_cnn, y_train,
                    validation_data=(X_val_cnn, y_val),
                    epochs=50, batch_size=32, callbacks=[early_stop], verbose=1)

# Evaluation
y_pred_cnn_frac = model.predict(X_test_cnn)
y_pred_cnn = y_pred_cnn_frac * 100

mae_cnn = [mean_absolute_error(y_test_perc[:, i], y_pred_cnn[:, i]) for i in range(3)]
print(f"CNN MAE (%): BW={mae_cnn[0]:.2f}, MA={mae_cnn[1]:.2f}, EM={mae_cnn[2]:.2f} | Avg={np.mean(mae_cnn):.2f}")

# Derived multi-label classification metrics (threshold 10% for presence)
THRESH = 0.10
y_test_bin = (y_test > THRESH).astype(int)
y_pred_bin_cnn = (y_pred_cnn_frac > THRESH).astype(int)
y_pred_bin_svm = (y_pred_svm_frac > THRESH).astype(int)

print("\nDerived Multi-label Metrics (@10% threshold):")
print(f"CNN  Hamming Loss: {hamming_loss(y_test_bin, y_pred_bin_cnn):.3f}")
print(f"CNN  Macro F1: {f1_score(y_test_bin, y_pred_bin_cnn, average='macro'):.3f}")
print(f"CNN  ROC-AUC (macro): {roc_auc_score(y_test_bin, y_pred_cnn_frac, average='macro'):.3f}")
print(f"SVM  Hamming Loss: {hamming_loss(y_test_bin, y_pred_bin_svm):.3f}")
print(f"SVM  Macro F1: {f1_score(y_test_bin, y_pred_bin_svm, average='macro'):.3f}")
```

```python
# === 8. ANALYZE FUNCTION (New Format) ===
def analyze_ecg_segment(signal: np.ndarray):
    if len(signal) != SEG_LEN:
        raise ValueError(f"Signal must be {SEG_LEN} samples long.")

    seg_norm = min_max_normalize(signal)

    # CNN prediction
    seg_cnn = seg_norm[np.newaxis, ..., np.newaxis]
    pred_cnn_frac = model.predict(seg_cnn, verbose=0).flatten()

    # SVM prediction
    feat = extract_features(seg_norm)[np.newaxis, :]
    feat_scaled = scaler.transform(feat)
    pred_svm_frac = svr.predict(feat_scaled).flatten()

    # Ensemble average
    pred_frac = (pred_cnn_frac + pred_svm_frac) / 2
    pred_perc = pred_frac * 100

    scores = {
        'baseline_wander': round(float(pred_perc[0]), 1),
        'muscle_artifact': round(float(pred_perc[1]), 1),
        'electrode_motion': round(float(pred_perc[2]), 1)
    }

    detected = {k: v > 10.0 for k, v in scores.items()}
    severity_scores = {k: v for k, v in scores.items() if detected[k]}

    total_noise_perc = sum(scores.values())

    if any(detected.values()):
        dominant_noise = max(severity_scores, key=severity_scores.get).replace('_', ' ')
        dominant_noise = dominant_noise.replace('baseline wander', 'baseline_wander')  # match key style
    else:
        dominant_noise = 'none'

    # Overall quality category (tunable)
    if total_noise_perc < 20:
        overall_quality = 'excellent'
    elif total_noise_perc < 40:
        overall_quality = 'good'
    elif total_noise_perc < 60:
        overall_quality = 'moderate'
    else:
        overall_quality = 'poor'

    return {
        'detected_noises': {
            'baseline_wander': detected['baseline_wander'],
            'muscle_artifact': detected['muscle_artifact'],
            'electrode_motion': detected['electrode_motion']
        },
        'severity_scores': severity_scores,
        'dominant_noise': dominant_noise,
        'overall_quality': overall_quality
    }
```

```python
# === 9. TEST THE FUNCTION ===
print("=== EXAMPLE ANALYSES ===")
for _ in range(5):
    idx = random.randint(0, len(X_test) - 1)
    test_signal = (X_test[idx] * 2) - 1  # rough denormalize for realistic amplitudes

    result = analyze_ecg_segment(test_signal)
    print("\nResult:")
    print(result)

    plt.figure(figsize=(10, 3))
    plt.plot(test_signal)
    plt.title(f"Segment - Quality: {result['overall_quality']} | Dominant: {result['dominant_noise']}")
    plt.grid(True)
    plt.show()
```
