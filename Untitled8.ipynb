```python
class DataSynthesizer:
    """Synthesize mixed ECG signals with controlled noise severity via SNR"""

    def __init__(self, clean_segments, bw_noise_full, ma_noise_full, em_noise_full):
        self.clean_segments = clean_segments
        self.bw_noise = bw_noise_full
        self.ma_noise = ma_noise_full
        self.em_noise = em_noise_full
        self.segment_length = 1800

    def _sample_noise_snippet(self, noise_full):
        start = random.randint(0, len(noise_full) - self.segment_length)
        return noise_full[start:start + self.segment_length]

    def _add_noise_at_snr(self, clean, noise_snip, snr_db):
        if snr_db == float('inf'):  # No noise
            return clean, 0.0
        p_clean = np.mean(clean ** 2)
        p_noise = np.mean(noise_snip ** 2)
        if p_noise == 0:
            return clean, 0.0
        scale = np.sqrt(p_clean / (p_noise * 10 ** (snr_db / 10.0)))
        noisy = clean + scale * noise_snip
        added_power = np.mean((scale * noise_snip) ** 2)
        return noisy, added_power

    def generate_dataset(self, n_samples=6000, snr_range=(-10, 20)):
        X = []
        y = []  # fractions [BW, MA, EM]

        for _ in range(n_samples):
            clean = random.choice(self.clean_segments)
            mixed = clean.copy()
            added_powers = [0.0, 0.0, 0.0]

            # Random number of noise types (0-3)
            num_noises = random.randint(0, 3)
            noise_indices = random.sample(range(3), num_noises) if num_noises > 0 else []

            for idx in noise_indices:
                if idx == 0:  # BW
                    snip = self._sample_noise_snippet(self.bw_noise)
                elif idx == 1:  # MA
                    snip = self._sample_noise_snippet(self.ma_noise)
                else:  # EM
                    snip = self._sample_noise_snippet(self.em_noise)

                snr_db = random.uniform(*snr_range)
                mixed, added_p = self._add_noise_at_snr(mixed, snip, snr_db)
                added_powers[idx] += added_p  # Accumulate (approximate for overlapping power)

            total_power = np.mean(mixed ** 2) if np.mean(mixed ** 2) > 0 else 1e-10
            fractions = [p / total_power for p in added_powers]

            X.append(mixed)
            y.append(fractions)

        return np.array(X), np.array(y)
```

```python
class Preprocessor:
    """Preprocess ECG signals for model training"""

    def __init__(self, fs=360):
        self.fs = fs
        self.scaler = MinMaxScaler(feature_range=(0, 1))

    def normalize(self, signals):
        """Normalize signals to [0, 1] range"""
        # Reshape for scaler
        original_shape = signals.shape
        signals_reshaped = signals.reshape(-1, 1)

        # Normalize
        normalized = self.scaler.fit_transform(signals_reshaped)

        # Reshape back
        return normalized.reshape(original_shape)

    def prepare_for_cnn(self, signals):
        """Prepare signals for CNN input (n_samples, 1800, 1)"""
        return signals.reshape(-1, signals.shape[1], 1)

    def prepare_for_svm(self, signals):
        """Keep signals as 2D for feature extraction"""
        return signals
```

```python
class FeatureExtractor:
    """Extract features from ECG signals for SVM"""

    def __init__(self):
        self.wavelet = 'db6'      # Common choice for ECG (you can change to 'db4' if preferred)
        self.level = 5            # Good resolution for noise frequencies

    def extract_dwt_features(self, signal):
        """
        Extract Discrete Wavelet Transform features

        Returns:
            features: Energy and std of approximation and detail coefficients
        """
        coeffs = pywt.wavedec(signal, self.wavelet, level=self.level)

        features = []
        for coeff in coeffs:
            # Energy
            energy = np.sum(coeff ** 2)
            features.append(energy)

            # Standard deviation
            std = np.std(coeff)
            features.append(std)

        return np.array(features)

    def extract_terma_features(self, signal):
        """
        Extract TERMA (Two Event-Related Moving Averages) features

        This method distinguishes QRS complexes from noise
        """
        # Square the signal
        squared_signal = signal ** 2

        # Short window moving average (QRS detection)
        window_short = 20 # samples (~55 ms at 360 Hz)
        ma_short = np.convolve(squared_signal, np.ones(window_short) / window_short, mode='same')

        # Long window moving average (baseline)
        window_long = 100 # samples (~278 ms at 360 Hz)
        ma_long = np.convolve(squared_signal, np.ones(window_long) / window_long, mode='same')

        # TERMA signal
        terma_signal = ma_short - ma_long

        # Extract features from TERMA signal
        features = []

        # Peak count (robust peak detection)
        peaks, _ = find_peaks(terma_signal, prominence=0.5 * np.std(terma_signal))
        features.append(len(peaks))

        # Mean amplitude
        features.append(np.mean(np.abs(terma_signal)))

        # Standard deviation
        features.append(np.std(terma_signal))

        return np.array(features)

    def extract_features(self, signals):
        """
        Extract all features from multiple signals

        Args:
            signals: (n_samples, 1800)

        Returns:
            features: (n_samples, n_features) ~13-20 features depending on DWT level
        """
        all_features = []

        for signal in signals:
            dwt_feats = self.extract_dwt_features(signal)
            terma_feats = self.extract_terma_features(signal)

            # Concatenate all features
            combined = np.concatenate([dwt_feats, terma_feats])
            all_features.append(combined)

        return np.array(all_features)
```

```python
class SVMNoiseDetector:
    """SVM-based noise detection and severity estimation"""

    def __init__(self):
        self.model = None
        self.feature_extractor = FeatureExtractor()

    def train(self, X_train, y_train, X_val, y_val):
        """
        Train SVM model

        Args:
            X_train: (n_samples, 1800) raw signals
            y_train: (n_samples, 3) noise fractions [BW, MA, EM]
        """
        print("Extracting features for SVM training...")
        X_train_features = self.feature_extractor.extract_features(X_train)
        X_val_features = self.feature_extractor.extract_features(X_val)

        print(f"Feature shape: {X_train_features.shape}")

        # Create multi-output SVR model
        base_svr = SVR(
            kernel='rbf',
            C=10.0,
            gamma='scale',
            epsilon=0.01
        )

        self.model = MultiOutputRegressor(base_svr, n_jobs=-1)

        print("Training SVM model...")
        self.model.fit(X_train_features, y_train)

        # Validation performance
        y_val_pred = self.model.predict(X_val_features)
        val_mae = mean_absolute_error(y_val, y_val_pred)
        print(f"Validation MAE: {val_mae:.4f}")

        return self.model

    def predict(self, X):
        """Predict noise fractions"""
        X_features = self.feature_extractor.extract_features(X)
        predictions = self.model.predict(X_features)

        # Clip predictions to [0, 1]
        predictions = np.clip(predictions, 0, 1)

        return predictions

class CNNNoiseDetector:
    """CNN-based noise detection and severity estimation"""

    def __init__(self):
        self.model = None
        self.history = None

    def build_model(self, input_shape=(1800, 1)):
        """Build 1D-CNN architecture"""
        inputs = Input(shape=input_shape)

        # First convolutional block
        x = Conv1D(32, 7, activation='relu', padding='same')(inputs)
        x = MaxPooling1D(2)(x)
        x = BatchNormalization()(x)

        # Second convolutional block
        x = Conv1D(64, 5, activation='relu', padding='same')(x)
        x = MaxPooling1D(2)(x)
        x = BatchNormalization()(x)

        # Third convolutional block
        x = Conv1D(128, 3, activation='relu', padding='same')(x)
        x = GlobalAveragePooling1D()(x)

        # Fully connected layers
        x = Dense(64, activation='relu')(x)
        x = Dropout(0.3)(x)

        x = Dense(32, activation='relu')(x)
        x = Dropout(0.2)(x)

        # Output layer (3 nodes for BW, MA, EM fractions)
        outputs = Dense(3, activation='sigmoid')(x)

        model = Model(inputs, outputs)

        model.compile(
            optimizer=Adam(learning_rate=0.001),
            loss='mse',
            metrics=['mae']
        )

        self.model = model
        return model

    def train(self, X_train, y_train, X_val, y_val, epochs=50, batch_size=32):
        """
        Train CNN model

        Args:
            X_train: (n_samples, 1800, 1) raw signals
            y_train: (n_samples, 3) noise fractions
        """
        if self.model is None:
            self.build_model()

        # Callbacks
        early_stopping = EarlyStopping(
            monitor='val_loss',
            patience=10,
            restore_best_weights=True,
            verbose=1
        )

        reduce_lr = ReduceLROnPlateau(
            monitor='val_loss',
            factor=0.5,
            patience=5,
            min_lr=1e-6,
            verbose=1
        )

        print("Training CNN model...")
        self.history = self.model.fit(
            X_train, y_train,
            validation_data=(X_val, y_val),
            epochs=epochs,
            batch_size=batch_size,
            callbacks=[early_stopping, reduce_lr],
            verbose=1
        )

        return self.model

    def predict(self, X):
        """Predict noise fractions"""
        predictions = self.model.predict(X, verbose=0)

        # Clip predictions to [0, 1]
        predictions = np.clip(predictions, 0, 1)

        return predictions


class ECGDataLoader:
    """Placeholder for ECGDataLoader class"""
    def __init__(self):
        pass

    def load_mit_bih_clean(self):
        # Dummy data for demonstration
        return np.random.rand(1000, 1800)

    def load_nstdb_noise(self, noise_type):
        # Dummy data for demonstration
        return np.random.rand(10000)

def analyze_ecg_segment(raw_signal, svm_detector, cnn_detector, preprocessor):
    """Placeholder for analyze_ecg_segment function"""
    # This function needs to be implemented to perform actual analysis
    # For now, it returns dummy results to avoid NameError

    # Preprocess the raw signal for the models
    processed_signal_svm = preprocessor.prepare_for_svm(raw_signal.reshape(1, -1))
    processed_signal_cnn = preprocessor.prepare_for_cnn(raw_signal.reshape(1, -1))

    # Predict noise fractions using SVM and CNN
    svm_predictions = svm_detector.predict(processed_signal_svm)[0]
    cnn_predictions = cnn_detector.predict(processed_signal_cnn)[0]

    # Average predictions (simple ensemble)
    ensemble_predictions = (svm_predictions + cnn_predictions) / 2

    noise_names = ['Baseline Wander', 'Muscle Artifact', 'Electrode Motion']

    # Determine dominant noise and overall quality
    if np.max(ensemble_predictions) > 0.1: # Threshold for significant noise
        dominant_noise_idx = np.argmax(ensemble_predictions)
        dominant_noise = noise_names[dominant_noise_idx]
        overall_quality = "Poor" # Or a more granular scale
    else:
        dominant_noise = "None"
        overall_quality = "Good"

    return {
        'svm_fractions': svm_predictions.tolist(),
        'cnn_fractions': cnn_predictions.tolist(),
        'ensemble_fractions': ensemble_predictions.tolist(),
        'dominant_noise': dominant_noise,
        'overall_quality': overall_quality
    }
```

```python
class ModelEvaluator:
    """Evaluate and compare model performance"""

    def __init__(self):
        self.noise_names = ['Baseline Wander', 'Muscle Artifact', 'Electrode Motion']

    def evaluate_model(self, y_true, y_pred, model_name, verbose=True):
        """Calculate comprehensive evaluation metrics"""
        if verbose:
            print(f"\n{'='*60}")
            print(f"{model_name} Performance Metrics")
            print(f"{'='*60}")

        # Overall metrics
        mae = mean_absolute_error(y_true, y_pred)
        rmse = np.sqrt(mean_squared_error(y_true, y_pred))
        r2 = r2_score(y_true, y_pred)

        if verbose:
            print(f"\nOverall Metrics:")
            print(f" Mean Absolute Error (MAE): {mae:.4f}")
            print(f" Root Mean Square Error (RMSE): {rmse:.4f}")
            print(f" R² Score: {r2:.4f}")

        # Per-noise-type metrics
        if verbose:
            print(f"\nPer-Noise-Type Metrics:")
        for i, noise_name in enumerate(self.noise_names):
            mae_noise = mean_absolute_error(y_true[:, i], y_pred[:, i])
            rmse_noise = np.sqrt(mean_squared_error(y_true[:, i], y_pred[:, i]))
            r2_noise = r2_score(y_true[:, i], y_pred[:, i])

            if verbose:
                print(f"\n {noise_name}:")
                print(f" MAE: {mae_noise:.4f}")
                print(f" RMSE: {rmse_noise:.4f}")
                print(f" R²: {r2_noise:.4f}")

        # Detection metrics (threshold at 10%)
        threshold = 0.10
        y_true_binary = (y_true > threshold).astype(int)
        y_pred_binary = (y_pred > threshold).astype(int)

        if verbose:
            print(f"\nDetection Metrics (threshold={threshold*100}%):")
        for i, noise_name in enumerate(self.noise_names):
            tp = np.sum((y_true_binary[:, i] == 1) & (y_pred_binary[:, i] == 1))
            fp = np.sum((y_true_binary[:, i] == 0) & (y_pred_binary[:, i] == 1))
            fn = np.sum((y_true_binary[:, i] == 1) & (y_pred_binary[:, i] == 0))
            tn = np.sum((y_true_binary[:, i] == 0) & (y_pred_binary[:, i] == 0))

            precision = tp / (tp + fp) if (tp + fp) > 0 else 0
            recall = tp / (tp + fn) if (tp + fn) > 0 else 0
            f1 = 2 * precision * recall / (precision + recall) if (precision + recall) > 0 else 0

            if verbose:
                print(f"\n {noise_name}:")
                print(f" Precision: {precision:.4f}")
                print(f" Recall: {recall:.4f}")
                print(f" F1-Score: {f1:.4f}")

        return {
            'mae': mae,
            'rmse': rmse,
            'r2': r2
        }

    def plot_results(self, y_true, y_pred_svm, y_pred_cnn, save_path='results'):
        """Generate comprehensive visualization plots"""
        import os
        os.makedirs(save_path, exist_ok=True)

        # 1. Scatter plots (Predicted vs Actual)
        fig, axes = plt.subplots(2, 3, figsize=(15, 10))
        fig.suptitle('Predicted vs Actual Noise Fractions', fontsize=16, fontweight='bold')

        for i, noise_name in enumerate(self.noise_names):
            # SVM
            axes[0, i].scatter(y_true[:, i], y_pred_svm[:, i], alpha=0.5, s=20)
            axes[0, i].plot([0, 1], [0, 1], 'r--', lw=2)
            axes[0, i].set_xlabel('True Fraction', fontsize=10)
            axes[0, i].set_ylabel('Predicted Fraction', fontsize=10)
            axes[0, i].set_title(f'SVM - {noise_name}', fontsize=11)
            axes[0, i].grid(True, alpha=0.3)

            # CNN
            axes[1, i].scatter(y_true[:, i], y_pred_cnn[:, i], alpha=0.5, s=20, color='orange')
            axes[1, i].plot([0, 1], [0, 1], 'r--', lw=2)
            axes[1, i].set_xlabel('True Fraction', fontsize=10)
            axes[1, i].set_ylabel('Predicted Fraction', fontsize=10)
            axes[1, i].set_title(f'CNN - {noise_name}', fontsize=11)
            axes[1, i].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.savefig(f'{save_path}/scatter_comparison.png', dpi=300, bbox_inches='tight')
        plt.close()

        # 2. Error distribution
        fig, axes = plt.subplots(2, 3, figsize=(15, 10))
        fig.suptitle('Prediction Error Distribution', fontsize=16, fontweight='bold')

        for i, noise_name in enumerate(self.noise_names):
            errors_svm = y_pred_svm[:, i] - y_true[:, i]
            errors_cnn = y_pred_cnn[:, i] - y_true[:, i]

            # SVM
            axes[0, i].hist(errors_svm, bins=50, alpha=0.7, color='blue', edgecolor='black')
            axes[0, i].axvline(0, color='r', linestyle='--', lw=2)
            axes[0, i].set_xlabel('Prediction Error', fontsize=10)
            axes[0, i].set_ylabel('Frequency', fontsize=10)
            axes[0, i].set_title(f'SVM - {noise_name}', fontsize=11)
            axes[0, i].grid(True, alpha=0.3)

            # CNN
            axes[1, i].hist(errors_cnn, bins=50, alpha=0.7, color='orange', edgecolor='black')
            axes[1, i].axvline(0, color='r', linestyle='--', lw=2)
            axes[1, i].set_xlabel('Prediction Error', fontsize=10)
            axes[1, i].set_ylabel('Frequency', fontsize=10)
            axes[1, i].set_title(f'CNN - {noise_name}', fontsize=11)
            axes[1, i].grid(True, alpha=0.3)

        plt.tight_layout()
        plt.savefig(f'{save_path}/error_distribution.png', dpi=300, bbox_inches='tight')
        plt.close()

        # 3. Model comparison bar chart
        metrics_svm = self.evaluate_model(y_true, y_pred_svm, "SVM", verbose=False)
        metrics_cnn = self.evaluate_model(y_true, y_pred_cnn, "CNN", verbose=False)

        fig, ax = plt.subplots(figsize=(10, 6))

        metrics_names = ['MAE', 'RMSE', 'R²']
        svm_values = [metrics_svm['mae'], metrics_svm['rmse'], metrics_svm['r2']]
        cnn_values = [metrics_cnn['mae'], metrics_cnn['rmse'], metrics_cnn['r2']]

        x = np.arange(len(metrics_names))
        width = 0.35

        ax.bar(x - width/2, svm_values, width, label='SVM', color='blue', alpha=0.8)
        ax.bar(x + width/2, cnn_values, width, label='CNN', color='orange', alpha=0.8)

        ax.set_xlabel('Metrics', fontsize=12, fontweight='bold')
        ax.set_ylabel('Value', fontsize=12, fontweight='bold')
        ax.set_title('Model Performance Comparison', fontsize=14, fontweight='bold')
        ax.set_xticks(x)
        ax.set_xticklabels(metrics_names)
        ax.legend()
        ax.grid(True, alpha=0.3)

        plt.tight_layout()
        plt.savefig(f'{save_path}/model_comparison_bar.png', dpi=300, bbox_inches='tight')
        plt.close()
```
